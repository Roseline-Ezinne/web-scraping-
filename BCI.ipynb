```python
#the url of each page chage by the value assigned to page
page = 0
# fetching relevant scrape results
Title = []
Available_as = []
Info = []
Others = []
# used a while loop to loop through each page
while page != 1000:
    url = f"https://scholar.google.com/scholar?start={page}&q=allintitle:+%22brain+computer+interface%22+&hl=en&as_sdt=0,5&as_ylo=2018&as_vis=1"
    
    headers ={'user-agent': "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/98.0.4758.102 Safari/537.36",
                'Accept-Language':'en-US,en;q=0.9',
              'Accept-Encoding': 'gzip, deflate, br',
              'Accept': 'text/html,application/xhtml+xml,application/xml;q=0.9,image/avif,image/webp,image/apng,*/*;q=0.8,application/signed-exchange;v=b3;q=0.9',
                'Referer': 'https://www.google.com/'
             }
    
    response = requests.get(url, headers = headers)
    # printed the response code to know if it is 200
    print(response)
    
    soup = bs(response.content, 'html.parser')
    
    contents = soup.find_all('div', class_='gs_r gs_or gs_scl')
    
    for c in contents:
        title = c.find('h3', class_='gs_rt').a.text

        Title.append(title)

        if c.find('div', class_='gs_ggs gs_fl') is not None:
            
            availability = c.find('span', class_='gs_ctg2').text
        
        else:
            
            availability = 'none'
        
        Available_as.append(availability)
        
        info = c.find('div', class_='gs_a').text
        
        Info.append(info)
        
        others = c.find('div', class_ = 'gs_fl').text
        
        Others.append(others)
        
        print('done')
    
    
    sleep(randint(10,30))
    
    page = page +  10


    
print('ALL_DONE!')
```
